# Sentiment Anylasis using Tensor Flow

```python
import pandas as pd 
df = pd.read_csv('trainingset.csv')
```

```python
review_df = df[['tweet', 'Pro Vaccine']]

##removing neutral reviews

review_df = review_df[review_df['Pro Vaccine'] != 0]
```

```python
review_df["Pro Vaccine"].value_counts()
```

```python
##convert the text into an array of vector embeddings

tweet = review_df.tweet.values
```

```python
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(num_words=5000)

tokenizer.fit_on_texts(tweet)
vocab_size = len(tokenizer.word_index) + 1



# fit_on_texts() method creates an association between the words and the assigned numbers
```

```python
encoded_docs = tokenizer.texts_to_sequences(tweet)
#replace the words with their assigned numbers using the text_to_sequence() 
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_sequence = pad_sequences(encoded_docs, maxlen=200)
```

## Tensor Flow Text Classifier

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense, Dropout, SpatialDropout1D
from tensorflow.keras.layers import Embedding

embedding_vector_length = 32
model = Sequential()
model.add(Embedding(vocab_size, embedding_vector_length, input_length=200))
model.add(SpatialDropout1D(0.25))
model.add(LSTM(50, dropout=0.5, recurrent_dropout=0.5))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
sentiment_label = review_df["Pro Vaccine"].factorize()
history = model.fit(padded_sequence,sentiment_label[0],validation_split=0.2, epochs=5, batch_size=10)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'], label='acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.legend()
plt.show()
plt.savefig("Accuracy plot.jpg")
```
